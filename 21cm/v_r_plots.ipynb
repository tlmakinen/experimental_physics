```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import scipy.stats
import pandas as pd
```

```python
# define known constants
R_o = 8.0   # distance from sun to center of galaxy
omega_o = 29 # km / s/ kpc
parameters = ['L', 'vPrime', 'v_lsr']
```

```python
data = []
for i in range(9):
    data_i = pd.read_csv('./galaxyHarvest/series002/velocities_002_0{}K.txt'.format(i+1), skiprows=0, sep=',')
    data_i.columns = parameters
    data.append(data_i)

data_10k = (pd.read_csv('./galaxyHarvest/series002/velocities_002_10K.txt', skiprows=0, sep=','))
data_10k.columns = parameters

data.append(data_10k)
```

```python
def radius(long, R_o):
    return R_o * np.sin(np.radians(long))

def velocity(long, R_o, omega_o, vlsr, vPrime):
    v_prime = vlsr - vPrime
    return v_prime + R_o * omega_o * np.sin(np.radians(long))
```

```python
r_vectors = []
v_vectors = []
for i in range(10):
    
    r_vectors.append(radius(data[i]['L'].values, R_o))
    v_vectors.append(velocity(data[i]['L'].values, R_o, omega_o, data[i]['v_lsr'], data[i]['vPrime']))
```

```python
# plot colors
colors = ['#1f77b4','#ff7f0e','#2ca02c','#d62728','#9467bd','#8c564b','#e377c2','#7f7f7f','#bcbd22','#17becf']
labels = ['1K', '2K', '3K', '4K', '5K', '6K', '7K', '8K', '9K', '10K']
```

```python
vprime_range = []
for i in range(len(data[1])):
    v_prime_at_r = []
    for table in data:
        v_prime_at_r.append(table['vPrime'][i])
    vprime_range.append(v_prime_at_r)
```

```python
vprime_std_dev = []
for i in vprime_range:
    vprime_std_dev.append(np.std(i))

del_vprime = vprime_std_dev

np.savetxt("del_vprime.csv", del_vprime, delimiter=",")
```

```python
len(del_vprime)
```

```python
# compile all the errors
del_R_o = 0.2
del_omega_o = 1.0
del_l = 0.1 # radians

def del_r(long, R_o, del_R_o, del_l):
    long = np.radians(long)
    return np.sqrt(((np.sin(long))*del_R_o)**2 + (R_o * np.cos(long) * del_l)**2)

def del_v(long, R_o, del_R_o, omega_o, del_omega_o, del_l, del_vprime):
    long = np.radians(long)
    return np.sqrt(np.asarray(del_vprime)**2 + (omega_o * np.sin(long) * del_R_o)**2 + (R_o * np.sin(long) * del_omega_o)**2 + (R_o * omega_o * np.cos(long) * del_l)**2)
```

```python
# compute r and v errors using our functions
r_err = []
v_err = []

for i in range(8):
    r_err.append(del_r(data[i]['L'].values, R_o, del_R_o, del_l))
    v_err.append(del_v(data[i]['L'].values, R_o, del_R_o, omega_o, del_omega_o, del_l, del_vprime))
```

```python
fig,ax = plt.subplots(1, 1, figsize=(15,9))

#for i in range(5):
    #ax.scatter(r_vectors[i+3], v_vectors[i+3], color = colors[i], marker = '+', label = labels[i+3])
    
ax.errorbar(r_vectors[6], v_vectors[6], xerr=r_err[6], yerr=v_err[6], color=colors[4], fmt = '^', label = '6K $v\' threshold')

plt.legend(fontsize=15, loc='best')
ax.set_xlabel('Galactic Radius (Kpc)', fontsize=22)
ax.set_ylabel('Rotational Velocity (km/s)', fontsize=22)

plt.savefig('plots/final_curve.png', dpi='figure')
```
